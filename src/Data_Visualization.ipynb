```python
%matplotlib notebook


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# A CSV file containing all scraped data from https://www.3ders.org/pricecompare/3dprinters/
complete_data = '../dataset/3d_printers.csv'
df = pd.read_csv(complete_data)
```

```python
complete_data = '../dataset/countries.csv'
df_country = pd.read_csv(complete_data)
```

```python

for i, country in df.iterrows():
    if country['Country'] == 'USA':
        df.at[i, 'Country'] = 'United States'
    elif country['Country'] == 'TaiWan':
        df.at[i, 'Country'] = 'Taiwan'
    elif country['Country'] == 'UK':
        df.at[i, 'Country'] = 'United Kingdom'
    elif country['Country'] == 'Korea':
        df.at[i, 'Country'] = 'South Korea'
    elif country['Country'] == 'Korean':
        df.at[i, 'Country'] = 'South Korea'
    elif country['Country'] == 'Italian':
        df.at[i, 'Country'] = 'Italy'
    elif country['Country'] == 'FRANCE':
        df.at[i, 'Country'] = 'France'
    elif country['Country'] == 'Barcelona':
        df.at[i, 'Country'] = 'Spain'
    elif country['Country'] == 'Argentina ':
        df.at[i, 'Country'] = 'Argentina'
    elif country['Country'] == ' contact':
        df.at[i, 'Country'] = 'Germany'


lat = []
long = []
for i, row in df_country.iterrows():
    for n, country in df.iterrows():
        if row['name'] == country['Country']:
            df.at[n, 'lat'] = row['latitude']
            df.at[n, 'long'] = row['longitude']

df.head()        
```

```python
file_name = ('../dataset/3d_printers_senvol.csv')
df_senvol = pd.read_csv(file_name)
```

```python
# Creating  duplicate Manufactures
df_unique_man = df[['Country', 'Manufacturer']]
df_unique_man = df_unique_man[~df_unique_man.duplicated(subset=['Manufacturer'])]
df_unique_man.head()
```

```python
# Adding Country to those manufactures that exist in our large dataset that exists in our newly created dataframe
for i, big_set in df_senvol.iterrows():
    for n, small_set in df_unique_man.iterrows():
        if small_set['Manufacturer'] == big_set['Manufacturer']:
            df_senvol.at[i, 'Country'] = df_unique_man.at[n, 'Country']
```

```python
# Combining our two datasets
df_result = pd.concat([df, df_senvol], join='outer', sort=False)
```

```python
# Removing duplicate Models
df_result = df_result[~df_result.duplicated(subset=['Model'])]
#df_result.to_csv('printers_all_data.csv', sep=',')
```

```python
# Removing rows with no Country value 
df_result_unique_country = df_result.dropna(subset=['Country'])
#df_result_unique_country.to_csv('printers_only_with_countries.csv', sep=',')
df_result_unique_country
```
