```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pygal
```

```python
# A CSV file containing all scraped data from https://www.3ders.org/pricecompare/3dprinters/
complete_data = '../dataset/3d_printers.csv'

df = pd.read_csv(complete_data)
df.head()
```

```python
complete_data = '../dataset/countries.csv'

df_country = pd.read_csv(complete_data)
df_country.head()
```

```python

for i, country in df.iterrows():
    if country['Country'] == 'USA':
        df.at[i, 'Country'] = 'United States'
    elif country['Country'] == 'TaiWan':
        df.at[i, 'Country'] = 'Taiwan'
    elif country['Country'] == 'UK':
        df.at[i, 'Country'] = 'United Kingdom'
    elif country['Country'] == 'Korea':
        df.at[i, 'Country'] = 'South Korea'
    elif country['Country'] == 'Korean':
        df.at[i, 'Country'] = 'South Korea'
    elif country['Country'] == 'Italian':
        df.at[i, 'Country'] = 'Italy'
    elif country['Country'] == 'FRANCE':
        df.at[i, 'Country'] = 'France'
    elif country['Country'] == 'Barcelona':
        df.at[i, 'Country'] = 'Spain'
    elif country['Country'] == 'Argentina ':
        df.at[i, 'Country'] = 'Argentina'
    elif country['Country'] == ' contact':
        df.at[i, 'Country'] = 'Germany'


lat = []
long = []
for i, row in df_country.iterrows():
    for n, country in df.iterrows():
        if row['name'] == country['Country']:
            df.at[n, 'lat'] = row['latitude']
            df.at[n, 'long'] = row['longitude']

df.head()        
```

```python
file_name = ('../dataset/3d_printers_senvol.csv')
df_senvol = pd.read_csv(file_name)
df_senvol
```

```python
df_unique_man = df[['Country', 'Manufacturer']]
df_unique_man = df_unique_man[~df_unique_man.duplicated(subset=['Manufacturer'])]
df_unique_man.head()
```

```python
for i, big_set in df_senvol.iterrows():
    for n, small_set in df_unique_man.iterrows():
        if small_set['Manufacturer'] == big_set['Manufacturer']:
            df_senvol.at[i, 'Country'] = df_unique_man.at[n, 'Country']

df_senvol
```

```python
df_result = pd.concat([df, df_senvol], join='outer', sort=False)
df_result
```

```python
df_result = df_result[~df_result.duplicated(subset=['Model'])]
df_result.to_csv('printers_all_data.csv', sep=',')
df_result
```

```python
df_result_unique_country = df_result.dropna(subset=['Country'])
df_result_unique_country.to_csv('printers_only_with_countries.csv', sep=',')
df_result_unique_country
```

```python
import pygal

for i, c in df_country.iterrows():
    for i, dc in df_result_unique_country.iterrows():
        if c['name'] == dc['Country']:
            df_result_unique_country.at[i, 'Country'] = c['country']
        

df_result_unique_country.to_csv('printers_unique_contry.csv', sep=',')
```

```python
worldmap_chart = pygal.maps.world.World()
worldmap_chart.title = 'Produced printers per country'

df_map = df_result_unique_country.Country.unique()
for countries in df_map:
    worldmap_chart.add(countries + ' ' + str(df_result_unique_country.Country.value_counts()[countries]), {
        countries.lower(): df_result_unique_country.Country.value_counts()[countries]
    })
    
worldmap_chart.render_in_browser()
```

```python
worldmap_chart = pygal.maps.world.World()
worldmap_chart.title = 'Heatmap of produced printers per country'

map_dict = {}

for countries in df_map:
    print(df_result_unique_country.Country.value_counts()[countries])
    map_dict.update({countries.lower(): df_result_unique_country.Country.value_counts()[countries]})

worldmap_chart.add('countries',
    map_dict
)
    
worldmap_chart.render_in_browser()
```

```python
line_chart = pygal.Bar()
line_chart.title = 'Printers per country'
for countries in df_map:
    line_chart.add(countries + ' ' + str(df_result_unique_country.Country.value_counts()[countries]),
                   df_result_unique_country.Country.value_counts()[countries])
line_chart.render_in_browser()
```

```python
line_chart = pygal.HorizontalBar()
line_chart.title = 'Printer price'
line_chart.add('< 49000', 19.5)
line_chart.add('50000 - 99999', 36.6)
line_chart.add('100000 - 249999', 36.3)
line_chart.add('250000 - 999999', 4.5)
line_chart.render()
```

```python
# Working with df_result which have combined our two datasets and removed duplicated models
df_combined_techs = df_result.copy()
df_combined_techs = df_combined_techs.reset_index(drop=True)
for i, tech in df_combined_techs.iterrows():
    isnan = pd.isnull(tech['Technology'])
    if isnan:
        df_combined_techs.at[i, 'Technology'] = df_combined_techs.at[i, 'AM Process']
        
df_combined_techs = df_combined_techs.dropna(subset=['Technology'])
df_combined_techs.to_csv('printers_combined_tech.csv', sep=',')
df_combined_techs
```

```python
line_chart = pygal.Bar()
line_chart.title = 'Techs used by 995 printers'
for tech in df_combined_techs.Technology.unique():
    line_chart.add(tech,
                   df_combined_techs.Technology.value_counts()[tech])
line_chart.render_in_browser()
```
